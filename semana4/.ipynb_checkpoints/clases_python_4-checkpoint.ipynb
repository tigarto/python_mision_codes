# **Matrices**

Una matriz es una estructura en forma de cuadrícula (un casillero rectangular) conformado por filas y columnas

![matriz](matriz.jpg)

Para referirnos a algún dato en el casillero debemos referenciarlo usando fila y columna. Asi, por ejemplo si deseamos acceder a un elemento que se encuentra en la fila **i** y en la columna **j** de una matrix llamada M, entonces la instrucción para acceder a este es:

```python
M[i][j]
```

````python
#%%tutor -l python3 -k
import random

class matriz:
    def __init__(self, m, n):
        self.m = m  # Numero de filas
        self.n = n  # Numero de columnas
        self.mat = []*(m + 1)
        for i in range(m + 1):
            a = [0]*(n + 1)
            self.mat.append(a)
            
    def construyeMatriz(self, r = 100):
        for i in range(1, self.m + 1):
            for j in range(1, self.n +1):
                self.mat[i][j] = random.randint(1, r)

    def imprimeMatrizPorFilas(self, mensaje = "Matriz sin nombre "):
        print("\n", mensaje)
        for i in range(1, self.m + 1):
            for j in range(1, self.n + 1):
                print(self.mat[i][j], "\t", end="")
            print()

    def imprimeMatrizPorColumnas(self, mensaje = "Matriz sin nombre "):
        print("\n", mensaje)
        for i in range(1, self.n + 1):
            for j in range(1, self.m + 1):
                print(self.mat[j][i], "\t", end="")
            print()
            
mat1 = matriz(4,5)
mat1.construyeMatriz(9)
mat1.imprimeMatrizPorFilas("mat1 (Por filas): ")
mat1.imprimeMatrizPorColumnas("mat1 (Por columnas): ")
````

````python
#%%tutor -l python3 -k
import random

class vector:
    def __init__(self, n):
        self.n = n
        self.V = [0] * (n + 1)
        
    def agregarDato(self, d):
        if self.esLleno():
            return
        self.V[0] = self.V[0] + 1
        self.V[self.V[0]] = d
    
    def esVacio(self):
        return self.V[0] == 0
    
    def esLleno(self):
        return self.V[0] == self.n
    
    def imprimeVector(self, mensaje="vector sin nombre: \t"):
        print("\n", mensaje, end=" ")
        for i in range(1, self.V[0] + 1):
            print(self.V[i], end=", ")
        print()

class matriz:
    def __init__(self, m, n):
        self.m = m  # Numero de filas
        self.n = n  # Numero de columnas
        self.mat = []*(m + 1)
        for i in range(m + 1):
            a = [0]*(n + 1)
            self.mat.append(a)
            
    def construyeMatriz(self, r = 100):
        for i in range(1, self.m + 1):
            for j in range(1, self.n +1):
                self.mat[i][j] = random.randint(1, r)

    def imprimeMatrizPorFilas(self, mensaje = "Matriz sin nombre "):
        print("\n", mensaje)
        for i in range(1, self.m + 1):
            for j in range(1, self.n + 1):
                print(self.mat[i][j], "\t", end="")
            print()

    def imprimeMatrizPorColumnas(self, mensaje = "Matriz sin nombre "):
        print("\n", mensaje)
        for i in range(1, self.n + 1):
            for j in range(1, self.m + 1):
                print(self.mat[j][i], "\t", end="")
            print()

    def sumarFilas(self):
        v = vector(self.m)
        for i in range(1, self.m + 1):
            s = 0
            for j in range(1, self.n + 1):
                s = s + self.mat[i][j]
            v.agregarDato(s)
        return v
    
    def sumarColumnas(self):
        v = vector(self.n)
        for i in range(1, self.n + 1):
            s = 0
            for j in range(1, self.m + 1):
                s = s + self.mat[j][i]
            v.agregarDato(s)
        return v
    
    def __add__(self, b):
        c = matriz(self.m, self.n)
        if self.m != b.m or self.n != b.n:
            print("Matrices no se pueden sumar. Son de dimensiones diferentes")
            return c
        for i in range(1, self.m + 1):
            for j in range(1, self.n + 1):
                c.mat[i][j] = self.mat[i][j] + b.mat[i][j]
        return c
    
    def productoDiagonalPrincipal(self):
        f = 1
        for i in range(1, self.m + 1):
            f = f * self.mat[i][i]
        return f
            
# Punto 1
A = matriz(4,4)
A.construyeMatriz(5)
B = matriz(4,4)
B.construyeMatriz(5)
# Punto 2
A.imprimeMatrizPorFilas("A: ")
B.imprimeMatrizPorFilas("B: ")
# Punto 3
suma_filas = A.sumarFilas()
suma_columnas = A.sumarColumnas()
suma_filas.imprimeVector("Suma de las filas:")
suma_columnas.imprimeVector("Suma de las filas:")
# Punto 4
C = A + B 
C.imprimeMatrizPorFilas("C: ")
# Punto 5
print("\nProducto diagonal principal: " + str(C.productoDiagonalPrincipal()))
````

````python
#%%tutor -l python3 -k
import random

class vector:
    def __init__(self, n):
        self.n = n
        self.V = [0] * (n + 1)
        
    def agregarDato(self, d):
        if self.esLleno():
            return
        self.V[0] = self.V[0] + 1
        self.V[self.V[0]] = d
    
    def esVacio(self):
        return self.V[0] == 0
    
    def esLleno(self):
        return self.V[0] == self.n
    
    def imprimeVector(self, mensaje="vector sin nombre: \t"):
        print("\n", mensaje, end=" ")
        for i in range(1, self.V[0] + 1):
            print(self.V[i], end=", ")
        print()

class matriz:
    def __init__(self, m, n):
        self.m = m  # Numero de filas
        self.n = n  # Numero de columnas
        self.mat = []*(m + 1)
        for i in range(m + 1):
            a = [0]*(n + 1)
            self.mat.append(a)
            
    def construyeMatriz(self, r = 100):
        for i in range(1, self.m + 1):
            for j in range(1, self.n +1):
                self.mat[i][j] = random.randint(1, r)

    def imprimeMatrizPorFilas(self, mensaje = "Matriz sin nombre "):
        print("\n", mensaje)
        for i in range(1, self.m + 1):
            for j in range(1, self.n + 1):
                print(self.mat[i][j], "\t", end="")
            print()

    def imprimeMatrizPorColumnas(self, mensaje = "Matriz sin nombre "):
        print("\n", mensaje)
        for i in range(1, self.n + 1):
            for j in range(1, self.m + 1):
                print(self.mat[j][i], "\t", end="")
            print()

    def sumarFilas(self):
        v = vector(self.m)
        for i in range(1, self.m + 1):
            s = 0
            for j in range(1, self.n + 1):
                s = s + self.mat[i][j]
            v.agregarDato(s)
        return v
    
    def sumarColumnas(self):
        v = vector(self.n)
        for i in range(1, self.n + 1):
            s = 0
            for j in range(1, self.m + 1):
                s = s + self.mat[j][i]
            v.agregarDato(s)
        return v
    
    def __add__(self, b):
        c = matriz(self.m, self.n)
        if self.m != b.m or self.n != b.n:
            print("Matrices no se pueden sumar. Son de dimensiones diferentes")
            return c
        for i in range(1, self.m + 1):
            for j in range(1, self.n + 1):
                c.mat[i][j] = self.mat[i][j] + b.mat[i][j]
        return c
    
    def productoDiagonalPrincipal(self):
        f = 1
        for i in range(1, self.m + 1):
            f = f * self.mat[i][i]
        return f

    def intercambiarFilas(self, i, j):
        for k in range(1, self.n):
            aux = self.mat[i][k]
            self.mat[i][k] = self.mat[j][k]
            self.mat[j][k] = aux
            
    def intercambiarColumnas(self, i, j):
        for k in range(1, self.m):
            aux = self.mat[k][i]
            self.mat[k][j] = self.mat[k][i]
            self.mat[k][i] = aux
            
    def traspuesta(self):
        c = matriz(self.n, self.m)
        for i in range(1, self.m + 1):
            for j in range(1, self.n + 1):
                c.mat[j][i] = self.mat[i][j]
        return c
    
# Punto 1
A = matriz(6,7)
A.construyeMatriz(9)
A.imprimeMatrizPorFilas("A: ")
# Punto 2
A.intercambiarFilas(2, 5)
A.imprimeMatrizPorFilas("A(F: 2 <--> 5): ")
# Punto 3
A.intercambiarColumnas(1, 7)
A.imprimeMatrizPorFilas("A(F: 2 <--> 5, C: 1 <---> 7): ")
# Punto 4
A_T = A.traspuesta()
A_T.imprimeMatrizPorFilas("A_T(F: 2 <--> 5, C: 1 <---> 7): ")
````

````python
#%%tutor -l python3 -k
import random

class vector:
    def __init__(self, n):
        self.n = n
        self.V = [0] * (n + 1)
        
    def agregarDato(self, d):
        if self.esLleno():
            return
        self.V[0] = self.V[0] + 1
        self.V[self.V[0]] = d
    
    def esVacio(self):
        return self.V[0] == 0
    
    def esLleno(self):
        return self.V[0] == self.n
    
    def imprimeVector(self, mensaje="vector sin nombre: \t"):
        print("\n", mensaje, end=" ")
        for i in range(1, self.V[0] + 1):
            print(self.V[i], end=", ")
        print()

class matriz:
    def __init__(self, m, n):
        self.m = m  # Numero de filas
        self.n = n  # Numero de columnas
        self.mat = []*(m + 1)
        for i in range(m + 1):
            a = [0]*(n + 1)
            self.mat.append(a)
            
    def construyeMatriz(self, r = 100):
        for i in range(1, self.m + 1):
            for j in range(1, self.n +1):
                self.mat[i][j] = random.randint(1, r)

    def imprimeMatrizPorFilas(self, mensaje = "Matriz sin nombre "):
        print("\n", mensaje)
        for i in range(1, self.m + 1):
            for j in range(1, self.n + 1):
                print(self.mat[i][j], "\t", end="")
            print()

    def imprimeMatrizPorColumnas(self, mensaje = "Matriz sin nombre "):
        print("\n", mensaje)
        for i in range(1, self.n + 1):
            for j in range(1, self.m + 1):
                print(self.mat[j][i], "\t", end="")
            print()

    def sumarFilas(self):
        v = vector(self.m)
        for i in range(1, self.m + 1):
            s = 0
            for j in range(1, self.n + 1):
                s = s + self.mat[i][j]
            v.agregarDato(s)
        return v
    
    def sumarColumnas(self):
        v = vector(self.n)
        for i in range(1, self.n + 1):
            s = 0
            for j in range(1, self.m + 1):
                s = s + self.mat[j][i]
            v.agregarDato(s)
        return v
    
    def __add__(self, b):
        c = matriz(self.m, self.n)
        if self.m != b.m or self.n != b.n:
            print("Matrices no se pueden sumar. Son de dimensiones diferentes")
            return c
        for i in range(1, self.m + 1):
            for j in range(1, self.n + 1):
                c.mat[i][j] = self.mat[i][j] + b.mat[i][j]
        return c
    
    def productoDiagonalPrincipal(self):
        f = 1
        for i in range(1, self.m + 1):
            f = f * self.mat[i][i]
        return f

    def intercambiarFilas(self, i, j):
        for k in range(1, self.n):
            aux = self.mat[i][k]
            self.mat[i][k] = self.mat[j][k]
            self.mat[j][k] = aux
            
    def intercambiarColumnas(self, i, j):
        for k in range(1, self.m):
            aux = self.mat[k][i]
            self.mat[k][j] = self.mat[k][i]
            self.mat[k][i] = aux
            
    def traspuesta(self):
        c = matriz(self.n, self.m)
        for i in range(1, self.m + 1):
            for j in range(1, self.n + 1):
                c.mat[j][i] = self.mat[i][j]
        return c

    def __mul__(self, b):
        c = matriz(self.m, b.n)
        if self.n != b.m:
            print("Matrices no se pueden multiplicar. ", end = "")
            print("el número de columnas de self es diferente del número de columnas de b")
            return c
        for i in range(1, self.m + 1):
            for j in range(1, b.n + 1):
                c.mat[i][j] = 0
                for k in range(1, self.n + 1):
                    c.mat[i][j] = c.mat[i][j] + self.mat[i][k] * b.mat[k][j]
        return c    
    
# Punto 1
A = matriz(3,4)
A.construyeMatriz(8)
B = matriz(4,3)
B.construyeMatriz(8)
C = A*B
# Punto 2
A.imprimeMatrizPorFilas("A: ")
B.imprimeMatrizPorFilas("B: ")
C.imprimeMatrizPorFilas("C = A x B: ")
````

````python
#%%tutor -l python3 -k
import random

class vector:
    def __init__(self, n):
        self.n = n
        self.V = [0] * (n + 1)
        
    def agregarDato(self, d):
        if self.esLleno():
            return
        self.V[0] = self.V[0] + 1
        self.V[self.V[0]] = d
    
    def esVacio(self):
        return self.V[0] == 0
    
    def esLleno(self):
        return self.V[0] == self.n
    
    def imprimeVector(self, mensaje="vector sin nombre: \t"):
        print("\n", mensaje, end=" ")
        for i in range(1, self.V[0] + 1):
            print(self.V[i], end=", ")
        print()

class matriz:
    def __init__(self, m, n):
        self.m = m  # Numero de filas
        self.n = n  # Numero de columnas
        self.mat = []*(m + 1)
        for i in range(m + 1):
            a = [0]*(n + 1)
            self.mat.append(a)
            
    def construyeMatriz(self, r = 100):
        for i in range(1, self.m + 1):
            for j in range(1, self.n +1):
                self.mat[i][j] = random.randint(1, r)

    def imprimeMatrizPorFilas(self, mensaje = "Matriz sin nombre "):
        print("\n", mensaje)
        for i in range(1, self.m + 1):
            for j in range(1, self.n + 1):
                print(self.mat[i][j], "\t", end="")
            print()

    def imprimeMatrizPorColumnas(self, mensaje = "Matriz sin nombre "):
        print("\n", mensaje)
        for i in range(1, self.n + 1):
            for j in range(1, self.m + 1):
                print(self.mat[j][i], "\t", end="")
            print()

    def sumarFilas(self):
        v = vector(self.m)
        for i in range(1, self.m + 1):
            s = 0
            for j in range(1, self.n + 1):
                s = s + self.mat[i][j]
            v.agregarDato(s)
        return v
    
    def sumarColumnas(self):
        v = vector(self.n)
        for i in range(1, self.n + 1):
            s = 0
            for j in range(1, self.m + 1):
                s = s + self.mat[j][i]
            v.agregarDato(s)
        return v
    
    def __add__(self, b):
        c = matriz(self.m, self.n)
        if self.m != b.m or self.n != b.n:
            print("Matrices no se pueden sumar. Son de dimensiones diferentes")
            return c
        for i in range(1, self.m + 1):
            for j in range(1, self.n + 1):
                c.mat[i][j] = self.mat[i][j] + b.mat[i][j]
        return c
    
    def productoDiagonalPrincipal(self):
        f = 1
        for i in range(1, self.m + 1):
            f = f * self.mat[i][i]
        return f

    def intercambiarFilas(self, i, j):
        for k in range(1, self.n):
            aux = self.mat[i][k]
            self.mat[i][k] = self.mat[j][k]
            self.mat[j][k] = aux
            
    def intercambiarColumnas(self, i, j):
        for k in range(1, self.m):
            aux = self.mat[k][i]
            self.mat[k][j] = self.mat[k][i]
            self.mat[k][i] = aux
            
    def traspuesta(self):
        c = matriz(self.n, self.m)
        for i in range(1, self.m + 1):
            for j in range(1, self.n + 1):
                c.mat[j][i] = self.mat[i][j]
        return c

    def __mul__(self, b):
        c = matriz(self.m, b.n)
        if self.n != b.m:
            print("Matrices no se pueden multiplicar. ", end = "")
            print("el número de columnas de self es diferente del número de columnas de b")
            return c
        for i in range(1, self.m + 1):
            for j in range(1, b.n + 1):
                c.mat[i][j] = 0
                for k in range(1, self.n + 1):
                    c.mat[i][j] = c.mat[i][j] + self.mat[i][k] * b.mat[k][j]
        return c    
    
    def sumaTriangularInferior(self):
        s = 0
        for i in range(1, self.m + 1):
            for j in range(1, i + 1):
                s = s + self.mat[i][j]
        return s

    def sumaTriangularEstrictamenteInferior(self):
        s = 0
        for i in range(2, self.m + 1):
            for j in range(1, i):
                s = s + self.mat[i][j]
        return s
    
    def sumaTriangularSuperior(self):
        s = 0
        for i in range(1, self.m + 1):
            for j in range(i, self.n + 1):
                s = s + self.mat[i][j]
        return s

    def sumaTriangularEstrictamenteSuperior(self):
        s = 0
        for i in range(1, self.m):
            for j in range(i + 1, self.n + 1):
                s = s + self.mat[i][j]
        return s    
    
# Punto 1
A = matriz(4,4)
A.construyeMatriz(4)
A.imprimeMatrizPorFilas("A: ")
# Punto 2
sum_triang_inf = A.sumaTriangularInferior()
print("Valor de la suma de los elementos de la triangular inferior:", sum_triang_inf)
# Punto 3
sum_triang_estrict_inf = A.sumaTriangularEstrictamenteInferior()
print("Valor de la suma de los elementos de la triangular estrictamente inferior:", sum_triang_estrict_inf)
# Punto 4
sum_triang_sup = A.sumaTriangularSuperior()
print("Valor de la suma de los elementos de la triangular superior:", sum_triang_sup)
# Punto 5
sum_triang_estrict_sup = A.sumaTriangularEstrictamenteSuperior()
print("Valor de la suma de los elementos de la triangular estrictamente superior:", sum_triang_estrict_sup)
````
